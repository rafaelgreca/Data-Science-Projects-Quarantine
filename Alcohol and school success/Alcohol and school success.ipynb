```python
#import libraries
import pandas as pd
pd.set_option('display.max_columns', 33)
```

```python
#open the data file
data = pd.read_csv("student-alcohol-consumption/student-mat.csv")

data.head(10)
```

```python
#check if the columns has any null value

data.isnull().sum()
```

```python
string_columns = data.select_dtypes(include = 'object')
string_columns
```

```python
#create a dummie column for all the string columns
data_copy = data.copy()

for column in string_columns:
    
    column_dummie = pd.get_dummies(data[column], drop_first = True)
    data_copy[column + '_' + column_dummie.columns] = column_dummie

#drop the string columns from the data
data_copy = data_copy.drop(string_columns, axis = 1)

column_names = data_copy.columns
column_names_reordered = ['school_MS', 'sex_M', 'age', 'address_U', 'famsize_LE3', 'Pstatus_T',
       'Medu', 'Fedu', 'Mjob_health', 'Mjob_other', 'Mjob_services', 'Mjob_teacher',
       'Fjob_health', 'Fjob_other', 'Fjob_services', 'Fjob_teacher',
       'reason_home', 'reason_other', 'reason_reputation', 'guardian_mother',
       'guardian_other', 'traveltime', 'studytime', 'failures','schoolsup_yes', 'famsup_yes', 'paid_yes',
       'activities_yes', 'nursery_yes', 'higher_yes', 'internet_yes',
       'romantic_yes', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2',
       'G3']

data_copy = data_copy[column_names_reordered]
data_copy
```

```python
data_copy.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns# Load dataset
sns.set(rc={'figure.figsize':(24.7,20.27)})

X = data_copy.drop(columns = ['G3'])
y = data_copy[['G3']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 19)

model  = RandomForestClassifier()
model.fit(X_train, y_train)

#check the most importants features using the Random Forest Classifier
importances = pd.Series(data=model.feature_importances_, index=X.columns)
sns.barplot(x=importances, y=importances.index, orient='h').set_title("Each feature importance")
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier

X = data_copy.drop(columns = ['G3'])
y = data_copy['G3']

#select the best features
embeded_rf_selector = SelectFromModel(RandomForestClassifier(n_estimators=100), max_features=30)
embeded_rf_selector.fit(X, y)

embeded_rf_support = embeded_rf_selector.get_support()
embeded_rf_feature = X.loc[:,embeded_rf_support].columns.tolist()
print(embeded_rf_feature)
print(str(len(embeded_rf_feature)), 'selected features')
```

## Conclusion: G2 is the variable that most affects the students grades. The Walc affects as well, but there is others variables that affect more (G1, G2, Absences, Health, Age, study time and so on..)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

best_columns_touse = ['age', 'Medu', 'Fedu', 'studytime', 'famrel', 'freetime', 'goout', 'Walc', 'health', 'absences', 'G1', 'G2']

X = data_copy[best_columns_touse]
y = data_copy['G3']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 19)

model = RandomForestClassifier()
model.fit(X_train, y_train)

model.score(X_test, y_test)
```
