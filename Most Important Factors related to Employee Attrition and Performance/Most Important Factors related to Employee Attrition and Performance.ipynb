```python
original_data = data.copy()

#map the binaries columns to 1 or 0
data['Attrition'] = data['Attrition'].map({'Yes': 1, 'No': 0})
data['Over18'] = data['Over18'].map({'Y': 1, 'N': 0})
```

```python
#create a dummie for every string column
string_columns = data.select_dtypes(include = 'object')

for column in string_columns:
    
    dummie = pd.get_dummies(data[column], drop_first = True)
    data[column + "_" + dummie.columns] = dummie

#delete the string columns from the dataset
data = data.drop(columns = string_columns)

data.head(10)
```

```python
#check the columns' types
data.dtypes
```

```python
#check if there is a nan value
data.isnull().sum()
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier

X = data.drop(columns = ['Attrition'])
y = data['Attrition']

#select the best features
embeded_rf_selector = SelectFromModel(RandomForestClassifier(n_estimators=100), max_features=30)
embeded_rf_selector.fit(X, y)

embeded_rf_support = embeded_rf_selector.get_support()
embeded_rf_feature = X.loc[:,embeded_rf_support].columns.tolist()
print(embeded_rf_feature)
```

```python
#analize the best features
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns# Load dataset
sns.set(rc={'figure.figsize':(24.7,20.27)})

X = data.drop(columns = ['Attrition'])
y = data['Attrition']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 19)

model  = RandomForestClassifier()
model.fit(X_train, y_train)

#check the most importants features using the Random Forest Classifier
importances = pd.DataFrame(data=model.feature_importances_, index=X.columns)

#order the dataframe
importances.sort_values(by = [0], ascending = False, inplace = True)
importances.plot(kind = 'bar')
#sns.barplot(x=importances, y=importances., orient='h').set_title("Each feature importance")
```
