## Find the weakest and strongest Pokemon

```python
#import libraries
import numpy as np
import pandas as pd

#disable max columns in pd options
pd.set_option('display.max_columns', None) 

#read the data
data = pd.read_csv("pokemon.csv")

data.describe()
```

```python
#used the describe function to get the values
max_attack = 185
min_attack = 5

strongest_pokemon_attack = []
weakest_pokemon_attack = []

strongest_pokemon_attack_gen = []
weakest_pokemon_attack_gen = []

strongest_pokemon_attack_type = []
weakest_pokemon_attack_type = []

#search for the pokemon who has the highest and lowest attack
#and save its type, name and generation
for rowns in range(data.shape[0]):
    if data['attack'][rowns] == max_attack:
        strongest_pokemon_attack.append(data['name'][rowns])
        strongest_pokemon_attack_type.append(data['type1'][rowns])
        strongest_pokemon_attack_gen.append(data['generation'][rowns])
    
    if data['attack'][rowns] == min_attack:
        weakest_pokemon_attack.append(data['name'][rowns])
        weakest_pokemon_attack_type.append(data['type1'][rowns])
        weakest_pokemon_attack_gen.append(data['generation'][rowns])
        
print("Using the atribute attack, the strongest Pokemon is: " + str(strongest_pokemon_attack) + " from generation " + str(strongest_pokemon_attack_gen) + " and type " + str(strongest_pokemon_attack_type))
print("Using the atribute attack, the weakest Pokemons are: " + str(weakest_pokemon_attack) + " from generation " + str(weakest_pokemon_attack_gen) + " respectively and types " + str(weakest_pokemon_attack_type))
```

## Legendary Pokemon Classifier

```python
#check if there is any columns with nan values
data.info()
```

```python
from sklearn.impute import SimpleImputer
from sklearn import preprocessing# Get column names first

#exclude the object (string) columns from the data set
data = data.select_dtypes(exclude = 'object')

#fill the nan values with the mean of each column
data['weight_kg'].fillna(data['weight_kg'].mean(), inplace = True)
data['height_m'].fillna(data['height_m'].mean(), inplace = True)
data['percentage_male'].fillna(data['percentage_male'].mean(), inplace = True)

names = data.columns# Create the Scaler object

#use the standardscaler to standardize the data
scaler = preprocessing.StandardScaler()# Fit your data on the scaler object
scaled_data = scaler.fit_transform(data)
scaled_data = pd.DataFrame(scaled_data, columns=names)

#show the scaled data
scaled_data
```

```python
#checking if all the columns were filled and there is no nan value left
scaled_data.info()
```

```python
#check the correlation between the atributes from the data
scaled_data.corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

data_copy = scaled_data.copy()

#best columns to predicte if a pokemon is legendary
#the criteria was to pick the 16 columns with the highest correlation value with the "is_legendary" column
best_columns = ['generation', 'weight_kg', 'speed', 'sp_defense', 'sp_attack', 'pokedex_number', 'hp', 'height_m',
                'experience_growth', 'defense', 'base_total', 'base_happiness', 'base_egg_steps', 'attack', 'against_ghost',
                'against_dark']

X = data_copy[best_columns]

#target
y = data_copy['is_legendary']

#split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 39)

#create the randomforestregressor and fit the model
clf = RandomForestRegressor(n_estimators = 50, max_depth = 17, random_state = 40)
clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)
print(score)
```
